# Fill the missing values of the data frame

Fill the missing values from the data frame using fillna ()

fill values using ffill(), bfill(), interpolate() - choose values between upper and lower rows


```python
# @title
df['Married'] = df['Married'].fillna('Yes')
```

```python
gender_count = df.groupby(['Gender','Dependents']).size()
gender_count
```

```python
for column in df.columns:
  value_counts = df[column].value_counts()
  print(f"Value counts for column '{column}':\n{value_counts}\n")
```

```python
df['Gender'] = df['Gender'].fillna('Male')
df['Dependents'] = df['Dependents'].fillna('0')
df['Self_Employed'] = df['Self_Employed'].fillna('No')
df['LoanAmount'] = df['LoanAmount'].fillna(df['LoanAmount'].mean())
df['Loan_Amount_Term'] = df['Loan_Amount_Term'].fillna(df['Loan_Amount_Term'].mean())
df['Credit_History'] = df['Credit_History'].fillna(df['Credit_History'].mean())
```

```python
df = df.drop(columns=['Loan_ID'])
df
```

remove duplicate values from the table

```python
df.drop_duplicates()
df
```

# Encoding

**Encoding**

**One hot encoding** - One-hot encoding assigns a binary value to each category in the categorical
column. : Data diamention increase exponentially but keeps the same weight among data set


```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder()
encoded_data = encoder.fit_transform(df[['Property_Area']])
encoded_df = pd.DataFrame(encoded_data.toarray(), columns=encoder.get_feature_names_out(['Property_Area']))
onehot_df = pd.concat([df, encoded_df], axis=1)
onehot_df
```

***Label Encoding*** - Label encoding assigns a numerical value to each category in the categorical
column

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
label_df = df
label_df['Property_Area'] = label_encoder.fit_transform(label_df['Property_Area'])
label_df
```
